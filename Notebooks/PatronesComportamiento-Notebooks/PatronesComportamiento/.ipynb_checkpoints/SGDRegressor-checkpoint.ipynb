Las siglas de SGD provienen de Decenso Estocástico del Gradiente: el gradiende e la pérdida es estimado con cada ejemplo y el moelo es actualizado  a lo larog del camino con un programade fuerza decreciente, también llamado tasa de aprendizaje.

```python
import pickle
from sklearn.linear_model import SGDRegressor
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_absolute_percentage_error

from sklearn.preprocessing import StandardScaler
```

Importamos los ficheros de entrenaimento, test y validación.

```python
with open('/content/gdrive/MyDrive/PatronesComportamiento/trainingX.pkl', 'rb') as file:
        X_train=pickle.load(file)
with open('/content/gdrive/MyDrive/PatronesComportamiento/trainingy.pkl', 'rb') as file:
        y_train=pickle.load(file)
with open('/content/gdrive/MyDrive/PatronesComportamiento/testX.pkl', 'rb') as file:
        X_test=pickle.load(file)
with open('/content/gdrive/MyDrive/PatronesComportamiento/testy.pkl', 'rb') as file:
        y_test=pickle.load(file)
with open('/content/gdrive/MyDrive/PatronesComportamiento/validateX.pkl', 'rb') as file:
        X_val=pickle.load(file)
with open('/content/gdrive/MyDrive/PatronesComportamiento/validatey.pkl', 'rb') as file:
        y_val=pickle.load(file)


with open('/content/gdrive/MyDrive/PatronesComportamiento/mySeriesInten.pkl', 'rb') as file:
        mySeries=pickle.load(file)
```

Aplanamos listas

```python
#aplanamos lista de listas
y_train2 = []
for lista in y_train:
    y_train2.extend(lista)
print(y_train2)
```

```python
#aplanamos lista de listas
y_test2 = []
for lista in y_test:
    y_test2.extend(lista)
print(y_test2)
```

Calculamos socre y predecimos un valor aislado.

```python

reg = make_pipeline(StandardScaler(),
                   SGDRegressor(max_iter=1000, tol=1e-3))
reg.fit(X_train, y_train2)
reg.score(X_train, y_train2)
```

Vamos a predecir los valores de test y calcular su MAPE error

```python
y_pred=[]
cont=0
for i in range(len(X_test)):
  y_pred.append(reg.predict([X_test.iloc[i]]))

```

```python
mean_absolute_percentage_error(y_test2, y_pred)
```

```python
n=123
real_value=y_test2[n]




print("Valor real: ",real_value)
print("Valor predicho: ",reg.predict([X_test.iloc[n]]))
```

Ahora vamos a predecir una serie entera, tomamos los w primeros valores de la serie y predecimos el w+1 y así sucesivamente.

```python
w=3
ser=500
lista=[]
i=0
while i < 288-w:
  frag=mySeries[ser][i:i+w]
  lista.extend(frag)
  lista.append(reg.predict([frag])[0])
  i=i+w+2


print(lista)
```
